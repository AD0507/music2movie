```python
#Check if the dataset is balanced?
#function to check balance and descriptive statistics of the dataset
def check_balance_revised(data):
    balance_info = {}
    for column in data.columns:
        if data[column].dtype == 'object' or data[column].dtype == 'bool':
            # For categorical columns, calculate value counts and percentage
            value_counts = data[column].value_counts(normalize=True) * 100
            balance_info[column] = {"type": "categorical", "balance": value_counts}
        elif pd.api.types.is_numeric_dtype(data[column]):
            # For numerical columns, calculate statistics
            min_val = data[column].min()
            max_val = data[column].max()
            mean_val = data[column].mean()
            std_dev = data[column].std()
            balance_info[column] = {
                "type": "numerical",
                "min": min_val,
                "max": max_val,
                "mean": mean_val,
                "std_dev": std_dev
            }
    return balance_info

# Apply the revised function to the Spotify dataset
balance_info_revised = check_balance_revised(spotify_data_final)

# Display the balance information for review
balance_info_revised
```

#What features most impact song popularity?

```python
#what features most impact song popularity?
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Prepare the data for modeling
features = spotify_data_final[
    ['danceability', 'energy', 'loudness', 'speechiness',
     'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
]
target = spotify_data_final['popularity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=45)

# Train a Random Forest Regressor to identify feature importance
model = RandomForestRegressor(n_estimators=100, random_state=45)
model.fit(X_train, y_train)

# Predict on the test set and calculate error
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

# Extract feature importances
feature_importances = pd.Series(model.feature_importances_, index=features.columns).sort_values(ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.values, y=feature_importances.index, palette='viridis')
plt.title("Feature Importances for Song Popularity")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.show()

# Display the feature importances and MSE
feature_importances, np.sqrt(mse)  # Returning RMSE to assess model performance
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Split the columns
final_df['genres_list'] = final_df['genres'].str.split(', ')
final_df['countries_list'] = final_df['production_countries'].str.split(', ')
final_df['languages_list'] = final_df['spoken_languages'].str.split(', ')

# 2. Explode each list column separately and reset the index
genres_exploded = final_df.explode('genres_list').reset_index(drop=True)
countries_exploded = final_df.explode('countries_list').reset_index(drop=True)
languages_exploded = final_df.explode('languages_list').reset_index(drop=True)

# 3. Combine the exploded columns into a single DataFrame
exploded_data = pd.DataFrame({
    'genres': genres_exploded['genres_list'],
    'countries': countries_exploded['countries_list'],
    'languages': languages_exploded['languages_list']
}).dropna()  # Drop NaNs after exploding

# 4. Define the function to calculate CramÃ©r's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# 5. Calculate the correlation coefficients using the function
cat_feats = ['genres', 'countries', 'languages']
rows = []

for x in cat_feats:
    col = []
    for y in cat_feats:
        cramers = cramers_v(exploded_data[x], exploded_data[y])
        col.append(round(cramers, 2))
    rows.append(col)

# 6. Create a DataFrame from the results
cramers_results = np.array(rows)
df_corr = pd.DataFrame(cramers_results, columns=cat_feats, index=cat_feats)

# 7. Plot the heatmap using Seaborn
mypal_1 = ['#FC05FB', '#FEAEFE', '#FCD2FC', '#F3FEFA', '#B4FFE4', '#3FFEBA']

mask = np.triu(np.ones_like(df_corr, dtype=bool))
fig, ax = plt.subplots(figsize=(11, 6), facecolor=None)
cmap = sns.color_palette(mypal_1, as_cmap=True)

sns.heatmap(df_corr, mask=mask, cmap=cmap, vmax=1.0, vmin=0, center=0, annot=True,
            square=False, linewidths=0.01, cbar_kws={"shrink": 0.75})

# Add title
ax.set_title("Categorical Features Correlation (CramÃ©r's V)", fontsize=15, y=1.05)

# Display the heatmap
plt.show()
```
